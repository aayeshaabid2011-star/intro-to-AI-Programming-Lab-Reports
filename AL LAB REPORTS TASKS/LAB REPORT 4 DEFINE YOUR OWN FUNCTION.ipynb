### In circuit,you often calculate series resistances.
•	Write a function series() that takes list or tuple of resistances.

•	Return the total resistance.

•	Add docstrings and annotations.

•	Print the function’s __annotations__.circuit


```python
def series(resistances: list) -> float:
    
    total = 0
    for r in resistances:
        total += r
    return total

resistors = [10, 20, 90]
total_resistance = series(resistors)
print("Total Series Resistance:", total_resistance, "Ω")
print("Function annotations:", series.__annotations__)
```
